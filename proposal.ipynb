## Gender-based differences in mathematical abilities

Byrnes (2005, p. 74) lists the following three main observations concerning the performance gap in mathematics:
- Male students perform much better than female students on the SAT math test.
- There is no convincing evidence for such gender differences prior to high school.
- The subjective assessment of their own competence is higher in case of male students in both elementary and high school - independent of actual abilities.
    
    
Nutall et. al. (2005) highlight that this male superiority in mathematical skills is only prominent among high-ability students in middle or higher schools, and is mainly reflected in the results of achievement tests (such as the SATs), while female students in general get better grades, even in mathematical subjects.


A US-wide study including the performance of students from grade 2 to 11 by Hide et. al. (2008) argues that these differences in performance have completely disappeared by now. However, the authors note that standard tests lack more advanced tasks that could provide an in-depth assessment of mathematical abilities. Also, they found a greater variance in male students' performance, which might still explain the predominance of males in professions requiring highly advanced mathematical skills.

Brunner et. al. (2008) analysed the performance of german high-school students on several tests of cognitive abilities. They have found that the small observed difference in mathematical skills in favor of males becomes greater, if general cognitive abilities that facilitate task interpretation are controlled for. Thus, the lack of gender-based difference in the study by Hide et. al. might be due to the compensatory role of girls' verbal advantage.

## Spatial reasoning ability as a mediator

Mental rotation - the ability to rotate objects using one's visual imagination - shows a substantial gender difference. It is traditionally measured with the water levels test (WLT), in which subjects have to draw the surfice of water inside complex geometrical objects after an imaginary rotation. Nutall et. al. (2005) found a correlation between female students' mental rotation skills and their SAT math scores independent of their verbal abilities. Furthermore, previously significant gender differences completely disappeared after controlling for the ability to mentally rotate images. The authors also observed that when building with blocks, male high school students focus more on structural properties and are capable of building more stable constructions than female students.

Nutall et. al. remark that there's huge gender difference in the amount of interest and time spent building with blocks in childhood, which may severely affect the development of spatial reasoning skills. Boys are willing to fight to have the blocks for themselves, while girls are rarely interested in these toys. The first question is, to what extent is this difference in interests attributable to inborn traits, and to what extent to cultural factors.

## Neurobiological differences

Greater brain size, larger neurons, faster synapse growth and right hemispheric dominance in males may to some extent account for their advantage in mathematics, although evidence about such neurobiological differences is inconclusive. Even if there is a difference, it may not only be genetic or hormonal, but also dependent on experience: those, who engage their right hemisphere more often may consequently undergo additional brain development. (Byrnes, 2005)

Nutall et. al. (2005) used handedness as a proxy for contralateral hemisphere dominance, and identified three types:
- Left-handers have poorer verbal and varied spatial abilities.
- Homozygotic right-handers (with right-handed relatives) have stronger verbal and poorer spatial abilities.
- Heterozygotic right-handers (who have left-handed relatives as well) may have an advantage compared to homozygotics, as they carry both alleles and thus they are not as strongly left-hemisphere dominant.

The authors found that, within the math/science majors, heterozygotic right-handed women outperformed their homozygotic peers and achieved similarly to men in mental rotation tasks, even if controlled for parents’ educational level and math and verbal aptitude scores. In another study, for both boys and girls reporting extensive spatial experiences with their brothers in their childhood, heterozygotic right-handers performed significantly higher on the mental rotation test than their heterozygotic peers. These results indicate that handedness is a good proxy for the genetic component of differences in spatial and consequently mathematical abilities.

## Discouragement and preferential treatment

Early stimulation and the development of early skills affect later motivations and the exposure to tasks that further enhance abilities. More motivated and skillful students are more likely to get higher quality education and improve faster. Such differences snowball over time, leading to ever larger gaps in performance. These effects may have genetic as well as cultural confounders.

Stereotypes about typically masculine and typically feminine abilities may discourage girls and consequently, they might invest less in their mathematics studies from early on, be less likely to participate advanced math programmes, and perform worse due to the lack of confidence. However, there are some observations that are inconsistent with the discouragement argument. Amoung high-ability students, males and females only differ in terms of mathematics skills, but not in their expectations about their own performance. Also, female students in general get better grades than male students, and only perform worse in SAT-like tests. An additional effect of stereotypes may be that female students receive less attention from implicitly gender-biased teachers, who expect better performance from male students and therefore give them preferential treatment. This claim is supported by the observation that the severity of gender-related stereotypes and the difference in mathematical abilities are correlated across countries. (Byrnes, 2005)
