## Issue: gender-based differences in mathematical abilities

Byrnes (2005, p. 74) lists the following three main observations concerning the performance gap in mathematics:
- Male students perform much better than female students on the SAT math test.
- There is no convincing evidence for such gender differences prior to high school.
- The subjective assessment of their own competence is higher in case of male students in both elementary and high school - independent of actual abilities.
    
    
Nutall et. al. (2005) highlight that this male superiority in mathematical skills is only prominent among high-ability students in middle or higher schools, and is mainly reflected in the results of achievement tests (such as the SATs), while female students in general get better grades, even in mathematical subjects.


A US-wide study including the performance of students from grade 2 to 11 by Hide et. al. (2008) argues that these differences in performance have completely disappeared by now. However, the authors note that standard tests lack more advanced tasks that could provide an in-depth assessment of mathematical abilities. Also, they found a greater variance in male students' performance, which might still explain the predominance of males in professions requiring highly advanced mathematical skills.

Brunner et. al. (2008) analysed the performance of german high-school students on several tests of cognitive abilities. They have found that the small observed difference in math-related tasks in favor of males becomes greater, if general cognitive abilities that facilitate task interpretation are controlled for. Thus, in order to measure pure differences in mathematical skills we have to take into account the compensatory role of female students' verbal advantage.

### Spatial reasoning ability as a mediator

Mental rotation - the ability to rotate objects using one's visual imagination - shows a substantial gender difference. It is traditionally measured with the water levels test (WLT), in which subjects have to draw the surfice of water inside complex geometrical objects after an imaginary rotation. Nutall et. al. (2005) found a correlation between female students' mental rotation skills and their SAT math scores independent of their verbal abilities. Furthermore, previously significant gender differences completely disappeared after controlling for the ability to mentally rotate images. The authors also observed that when building with blocks, male high school students focus more on structural properties and are capable of building more stable constructions than female students. In light of these finding, I am going to use mental rotation ability as a proxy for early skills that are relevant to my query.

Nutall et. al. remark that there's huge gender difference in the amount of interest and time spent building with blocks in childhood, which may strongly affect the development of spatial reasoning skills. Boys are willing to fight to have the blocks for themselves, while girls are rarely interested in these toys. Thus, a preference for building block-type activities in childhood may be a good measure of early skill-enhancing  stimulation. In order to answer my query, I have to isolate the effect of cultural factors on these preferences from inborn traits that may also influence early skills directly and indirectly, though biologically determined interests.

### Genetic confounders

Differences in mental rotation ability are already detectable in 3- to 5-month-old infants, which suggests an inborn advantage of males that preceeds cultural influences. Furthermore, boys shortly after birth show stronger visual preferences for mechanical toys such as mobiles, balls and vehicles, while girls prefer toys with human attributes (e.g. dolls). (Alexander & Wilcox, 2012)

Nutall et. al. (2005) used handedness as a proxy for contralateral hemisphere dominance, and identified three types:
- Left-handers have poorer verbal and varied spatial abilities.
- Homozygotic right-handers (with right-handed relatives) have stronger verbal and poorer spatial abilities.
- Heterozygotic right-handers (who have left-handed relatives as well) may have an advantage compared to homozygotics, as they carry both alleles and thus they are not as strongly left-hemisphere dominant.

The authors found that, within the math/science majors, heterozygotic right-handed women outperformed their homozygotic peers and achieved similarly to men in mental rotation tasks, even if controlled for parents’ educational level and math and verbal aptitude scores. In another study, for both boys and girls reporting extensive spatial experiences with their brothers in their childhood, heterozygotic right-handers performed significantly higher on the mental rotation test than their heterozygotic peers. These results indicate that handedness is a good proxy for the genetic component of differences in spatial and consequently mathematical abilities. However, Byrnes (2005) warns that such hemispheric assymmetries may not only be of genetic or hormonal origin, but also dependent on experience: those, who engage their right hemisphere more often may consequently undergo additional brain development.

### Methods

Using Pearl & Mackenzie's (2018, p. 209) formula and the assumption that the only mediator of this effect is early stimulation, the estimand can be expressed as follows:

$\large P(Y | do(X)) = \sum_Z P(Z = z | X) \sum_X P(Y | X = x, Z = z) P(X = x)$

where $\large Y, X$ and $\large Z$ stand for early skills, gender and early stimulation, respectively. The problem is however that the confounder *genetics* may have a direct effect not only on the outcome, but also on the mediator, rendering frontdoor adjustment unfeasible. Therefore, I have to break the path into two further subpaths.

#### Genetics-independent effect of gender on early stimulation

Here I choose an RCT to avoid the most problematic backdoor path: the hardly measurable direct effect of genetics on early preferences. I assume that gender affects early stimulation mainly through children's tendency to follow the typical behavioral patterns of their peers. The experiment involves infants in the early phase of development, before nursery school that can be regarded as the first exposure to social norms. Before the treatment, participants are presented a forced-choice test between two types of toys: building blocks and a doll, which are meant to represent traditionally masculine mechanical-geometric and traditionally feminine verbal-social types of stimuli, respectively. The difference in the ratios ($\large D$) of boys and girls choosing the building blocks is measured.

Next, the children are randomly assigned to either of the following groups:
- In treatment group 1, subjects are repeatedly shown advertisement-like videos of boys playing with building blocks and girls playing with dolls, over the course of several weeks.
- In treatment group 2, the toy-gender pairings are reversed in the advertisements.
- In the control group, no videos are shown.

Then, the difference in the pre- and pos-treatment $\large D$ between the two treatment groups, as well as the treatment and control groups is calculated.

#### Effect of early stimulation on early skills

As any backdoor path between these two variables has to lead through inborn skills, which can be controlled for through handedness, I rely on observational data to estimate this effect. As early skills may lead to further advantages in elementary school, which belongs to the next section, this analysis involves a sample of nursery school children, for whom the following data are collected:
- $\large M$: performance on the mental rotation (WLT) task,
- $\large B$: standardized reports on the frequency of activities throughout nursery school that require geometrical thinking and mechanical skills (this is a composite index contructed of a few representative items, such as building blocks, vehicles, strategic computer games etc.),
- $\large H$: handedness of the subjects as well as their parents,
- $\large G$: gender,
- $\large PE$: parental education,
- $\large PI$: parental income.

The estimand is the effect of such skill-enhancing activities on mental rotation performance, conditional on the other factors: $\large \mathbb{E}_{H, G, PE, PI} [P(\textit{M } | \textit{ B, H, G, PE, PI})]$.

Gender is controlled for, because handedness might not entirely capture differences in inborn abilities. Variables $\large PE, PI$ are included, because and the composite index $\large B$ might disregard other types of cognitive stimuli confounded with early skills (e.g. wealthy, educated parents may have better genes and give more brain-teasers to their children).

## Subpath 2. - the effect of early skills on advanced skills

My second estimand is $\large P(\textit{advanced skills } | \textit{ do(early skills)})$. Early skills affect later motivations and the exposure to tasks that further enhance abilities. More motivated and skillful students are more likely to get higher quality education and improve faster. Such differences snowball over time, leading to ever larger gaps in performance.

## Discouragement and preferential treatment

Stereotypes about typically masculine and typically feminine abilities may discourage girls and consequently, they might invest less in their mathematics studies from early on, be less likely to participate advanced math programmes, and perform worse due to the lack of confidence. However, there are some observations that are inconsistent with the discouragement argument. Amoung high-ability students, males and females only differ in terms of mathematics skills, but not in their expectations about their own performance. Also, female students in general get better grades than male students, and only perform worse in SAT-like tests. An additional effect of stereotypes may be that female students receive less attention from implicitly gender-biased teachers and even parents, who expect better performance from male students and therefore give them preferential treatment. This claim is supported by the observation that the severity of gender-related stereotypes and the difference in mathematical abilities are correlated across countries. (Byrnes, 2005)
